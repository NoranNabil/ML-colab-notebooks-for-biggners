```python
import numpy as np
import cv2
from google.colab.patches import cv2_imshow

lena=cv2.imread('/content/lena.jpg')

cv2_imshow(lena)
```

```python
R=lena[:,:,0]
G=lena[:,:,1]
B=lena[:,:,2]

gray=np.zeros((1960,1960))
gray=0.3*R+0.6*G+0.1*B
cv2_imshow(gray)
```

```python
gray.shape
```

```python
U,D,V = np.linalg.svd(gray)


U=np.mat(U)
D=np.diag(D)
V=np.mat(V)

k=150
n=1960 #image height
m=1960 #image width
```

```python
U_=U[:,0:k]
D_=D[0:k,0:k]
V_=V[0:k,:]
```

```python
compressed_lena = U_*D_*V_
compressed_lena.shape

cv2_imshow(compressed_lena)
```

```python
comp_ratio=((n*k+k+k*m)/(n*m))*100
print(comp_ratio,"%")
```
